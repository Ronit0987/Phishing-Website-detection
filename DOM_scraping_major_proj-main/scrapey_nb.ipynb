```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import TimeoutException
import os
import pandas as pd
```

```python
df=pd.read_csv("F:\edge_downloads\phishing_urls.csv")
df=df.head(10)
df
```

```python
# Configure Chrome options
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run Chrome in headless mode (no GUI)
chrome_options.add_argument("--disable-gpu")  # Disable GPU acceleration

# Initialize Chrome WebDriver
driver = webdriver.Chrome(options=chrome_options)
driver.set_page_load_timeout(10)  # Timeout set to 10 seconds
# Create a directory to save DOM tree files
if not os.path.exists('dom_trees'):
    os.makedirs('dom_trees')

for index, row in df.iterrows():
    url = row['url']
    url_type = row['type']

    try:
        # Open the URL with a timeout
        driver.get(url)

        # Get the DOM tree
        dom_tree = driver.page_source

        # Save the DOM tree to a file
        filename = f"dom_trees/{url_type}_{url.replace('https://', '').replace('http://', '')}_dom_tree.html"
        with open(filename, 'w', encoding='utf-8') as file:
            file.write(dom_tree)

        print(f"DOM tree saved for {url} as {filename}")

    except TimeoutException:
        print(f"Timeout occurred while loading {url}. Proceeding to the next URL.")

# Quit the WebDriver
driver.quit()
```
