```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import TimeoutException, WebDriverException
import os
import pandas as pd

# Assuming 'df' is a pandas DataFrame with 'type' and 'url' columns
# Let's create a dummy 'df' for demonstration purposes


# Configure Chrome options
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run Chrome in headless mode (no GUI)
chrome_options.add_argument("--disable-gpu")  # Disable GPU acceleration

# Initialize Chrome WebDriver
driver = webdriver.Chrome(options=chrome_options)
driver.set_page_load_timeout(10)  # Timeout set to 10 seconds

# Create a directory to save DOM tree files
directory_name = 'F://React_projs2//DOM_scraping_major_proj//dom_trees_new_phish'
os.makedirs(directory_name, exist_ok=True)

# Add a 'size' column to the dataframe to store the size of the DOM tree
df['size'] = None

for index, row in df.iterrows():
    url = row['url'].rstrip('/')
    
    try:
        # Open the URL with a timeout
        driver.get(url)

        # Get the DOM tree
        dom_tree = driver.page_source

        # Save the DOM tree to a file
        filename = os.path.join(directory_name, f"{index+1}_dom_tree.html")  # Using index+1 for filename numbering
        with open(filename, 'w', encoding='utf-8') as file:
            file.write(dom_tree)

        # Update the 'size' column with the size of the DOM tree
        df.at[index, 'size'] = len(dom_tree)
        df.at[index, 'serial_no'] = index + 1
        print(f"DOM tree saved for {url} as {filename}")

    except (TimeoutException, WebDriverException) as e:
        print(f"Error occurred while loading {url}: {str(e)}. Proceeding to the next URL.")

# Quit the WebDriver
driver.quit()
# Print the updated dataframe
print(df)
```

```python
from bs4 import BeautifulSoup
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Function to read HTML content from a file
def read_html_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return file.read()

# Function to compute TF-IDF scores of HTML tags from a file
def tf_idf_tags(file_path):
    html_content = read_html_file(file_path)

    # Check if HTML content is empty
    if not html_content.strip():
        return {}  # Return an empty dictionary if content is empty

    # Parse the HTML content
    soup = BeautifulSoup(html_content, 'html.parser')

    # Extract tags from the DOM tree
    tags = [tag.name for tag in soup.find_all()]

    # Convert tags to string for TF-IDF computation
    tag_string = ' '.join(tags)

    # Compute TF-IDF scores
    tfidf = TfidfVectorizer()
    tfidf_matrix = tfidf.fit_transform([tag_string])

    # Get feature names (tags)
    feature_names = tfidf.get_feature_names_out()

    # Create a dictionary to store TF-IDF scores for each tag
    tag_tfidf_scores = {}
    for tag, score in zip(feature_names, tfidf_matrix.toarray()[0]):
        tag_tfidf_scores[tag] = score
    return tag_tfidf_scores

# Function to calculate similarity score between two documents based on their TF-IDF scores
def cal_similarity_score(idx):
    file_path1 = "F://React_projs2//DOM_scraping_major_proj//dom_trees_new_phish//" + str(idx) + "_dom_tree.html"
    tf_idf_tag = tf_idf_tags(file_path1)

    # Check if tf_idf_tag is empty
    if not tf_idf_tag:
        return 0  # Return 0 similarity score if tf_idf_tag is empty

    # Get tags from the reference document
    tags = set(res.keys()).union(set(tf_idf_tag.keys()))

    # Compute TF-IDF vectors for both documents
    vector_page1 = np.array([res.get(tag, 0) for tag in tags]).reshape(1, -1)
    vector_page2 = np.array([tf_idf_tag.get(tag, 0) for tag in tags]).reshape(1, -1)

    # Compute cosine similarity
    similarity_score = cosine_similarity(vector_page1, vector_page2)[0][0]
    return similarity_score

# File path of the reference HTML file

count=0
# Calculate TF-IDF scores for the reference HTML file
file_path="F://React_projs2//DOM_scraping_major_proj//dom_trees_new_phish//943_dom_tree.html"
res = tf_idf_tags(file_path)

# Print TF-IDF scores for each tag
for tag, score in sorted(res.items(), key=lambda x: x[1], reverse=True):
    print(f"{tag}: {score}")
df['similarity_score']=None
# Calculate and print similarity scores for other documents
for idx in df['serial_no']:
    similarity_score = cal_similarity_score(idx)
    df['similarity_score'].iloc[count]=similarity_score
    count=count+1
    print(f"Similarity score for document {idx}: {similarity_score}")
```

```python
from bs4 import BeautifulSoup
import requests
import re

# Function to extract features from HTML content
def extract_features(html_content):
    features = {}
    
    # Parse HTML content
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # Extract URL features
    url = soup.find('meta', attrs={'property': 'og:url'})
    #features['url_length'] = len(url['content']) if url else 0
    
    # Extract form features
    forms = soup.find_all('form')
    features['num_forms'] = len(forms)
    features['has_login_form'] = any(['login' in form.get('action', '').lower() for form in forms])
    
    # Extract script features
    scripts = soup.find_all('script')
    features['num_scripts'] = len(scripts)
    features['has_external_scripts'] = any(['http' in script.get('src', '') for script in scripts])
    
    # Extract hyperlink features
    hyperlinks = soup.find_all('a', href=True)
    features['num_hyperlinks'] = len(hyperlinks)
    features['num_external_hyperlinks'] = sum(1 for link in hyperlinks if 'http' in link['href'])
    
    # Extract JavaScript event features
    js_events = soup.find_all(re.compile('^on'))
    features['num_js_events'] = len(js_events)
    
    return features

# Example URL for testing
count=0
for idx in df['serial_no']:
    file_path1 = "F://React_projs2//DOM_scraping_major_proj//dom_trees_new_phish//" + str(idx) + "_dom_tree.html"
    html_content = read_html_file(file_path1)
    features=extract_features(html_content)
    df['num_forms'].iloc[count]=features['num_forms']
    df['has_login_form'].iloc[count]=features['has_login_form']
    df['num_scripts'].iloc[count]=features['num_scripts']
    df['has_external_scripts'].iloc[count]=features['has_external_scripts']
    df['num_hyperlinks'].iloc[count]=features['num_hyperlinks']
    df['num_external_hyperlinks'].iloc[count]=features['num_external_hyperlinks']
    df['num_js_events'].iloc[count]=features['num_js_events']
    count=count+1
    #print(features)

# # Fetch HTML content from the URL
# response = requests.get(url)
# html_content = response.text

# # Extract features from HTML content
# features = extract_features(html_content)

# # Print extracted features
# print("Extracted Features:")
# for key, value in features.items():
#     print(f"{key}: {value}")
```
